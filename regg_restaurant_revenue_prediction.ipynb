```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('train.csv.zip')
xtest=pd.read_csv('test.csv.zip')
```

```python
df.drop(columns=['Id'])
```

```python
df.columns
```

```python
df.describe()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
l=[i for i in df.columns if df[i].dtype=='object']
for i in l:
  df[i]= label_encoder.fit_transform(df[i])
```

```python
a=[i for i in xtest.columns if xtest[i].dtype=="object"]
for i in a:
  xtest[i]=label_encoder.fit_transform(xtest[i])
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
y
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
xtest=st.fit_transform(xtest)
xtest
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
# from sklearn.preprocessing import StandardScaler
sy=StandardScaler()
y=pd.DataFrame(y)
y=sy.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_valid,y_train,y_valid=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import Adam
model=Sequential()
model.add(Dense(units=25,input_dim=42,activation="tanh"))
model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=30,activation="relu"))
model.add(Dense(1))
```

```python
model.compile(loss='mse',optimizer=Adam(learning_rate=0.001),metrics=['RootMeanSquaredError'])
```

```python
model.summary()
```

```python

model.fit(x_train,y_train,epochs=50,validation_data=(x_valid,y_valid))
```

```python
mse,rmse = model.evaluate(x_train,y_train)
print(rmse)
```

```python
y_pred=model.predict(xtest)
```

```python
y_prd=sy.inverse_transform(y_pred)
```

```python
y_prd
```

```python
y_prd=pd.DataFrame(y_prd)
y_prd.to_csv('pred1.csv')
```
